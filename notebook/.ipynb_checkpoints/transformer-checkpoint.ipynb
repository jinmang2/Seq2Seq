```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
device = torch.device('cpu')
```

```python
SEQ_LEN     = 15
BATCH_SIZE  = 5
INPUT_DIM   = 30
OUTPUT_DIM  = 37
HID_DIM     = 256
ENC_LAYERS  = DEC_LAYERS  = 3
ENC_HEADS   = DEC_HEADS   = 8
ENC_PF_DIM  = DEC_PF_DIM  = 512
ENC_DROPOUT = DEC_DROPOUT = 0.1
```

```python
SRC_PAD_IDX = TRG_PAD_IDX = 1
MIN_WORDS   = 5

src_seq_length = torch.randint(MIN_WORDS, SEQ_LEN-1, (BATCH_SIZE,))
trg_seq_length = torch.randint(MIN_WORDS, SEQ_LEN-1, (BATCH_SIZE,))
if SEQ_LEN - 1 not in src_seq_length:
    src_seq_length[-1] = SEQ_LEN - 2
if SEQ_LEN - 1 not in trg_seq_length:
    trg_seq_length[-1] = SEQ_LEN - 2

x = torch.randint(0+2, INPUT_DIM-2, size=(BATCH_SIZE, SEQ_LEN))
x[:, 0] = 0
for i, ind in enumerate(src_seq_length):
    x[i, ind+1 ] = INPUT_DIM - 1
    x[i, ind+2:] = SRC_PAD_IDX

y = torch.randint(0+2, OUTPUT_DIM-2, size=(BATCH_SIZE, SEQ_LEN))
y[:, 0] = 0
for i, ind in enumerate(trg_seq_length):
    y[i, ind+1 ] = OUTPUT_DIM - 1
    y[i, ind+2:] = TRG_PAD_IDX

print(x, x.shape, end='\n\n')
print(y, y.shape)
```

```python
MAX_LENGTH = 100

tok_embedding = nn.Embedding(INPUT_DIM, HID_DIM)
pos_embedding = nn.Embedding(MAX_LENGTH, HID_DIM)
scale = torch.sqrt(torch.FloatTensor([HID_DIM]))
dropout = nn.Dropout(ENC_DROPOUT)
```

```python
tok_embedding.weight.shape, pos_embedding.weight.shape
```

- make src mask

```python
(x != SRC_PAD_IDX)
```

```python
src_mask = (x != SRC_PAD_IDX).unsqueeze(1).unsqueeze(2)
src_mask.shape # 나중에 energy input size가 (bsz, n_heads, Q seq_len, K seq_len)이 됨.
               # key부분의 padding 값을 -1e10으로 채울 것임
```

- make trg mask

```python
(y != TRG_PAD_IDX)
```

```python
trg_pad_mask = (y != SRC_PAD_IDX).unsqueeze(1).unsqueeze(3)
trg_pad_mask.shape
```

```python
trg_len = y.shape[1]
trg_len
```

```python
trg_sub_mask = torch.tril(
    torch.ones((trg_len, trg_len), device=device)).bool()
trg_sub_mask
```

```python
trg_mask = trg_pad_mask & trg_sub_mask
```

```python
trg_mask.shape
```

```python
trg_mask.shape
```

```python
x.shape
```

```python
# Positional Embedding을 위한 position값 생성
pos = torch.arange(0, SEQ_LEN).unsqueeze(0).repeat(BATCH_SIZE, 1)
pos, pos.shape
```

```python
src = dropout(tok_embedding(x) * scale + pos_embedding(pos))
# x's shape is   (bsz, seq_len): token 정보
# pos's shape is (baz, seq_len): position 정보
# tok_emb: x => emb(x) \in \mathbb{R}^{HID_DIM}
# x는 bsz * seq_len만큼의 원소를 가지고있음!
# x \in \mathbb{Z}_{set of range(0, INPUT_DIM)}
# 이를 emb(x)로 \mathbb{R}^{HID_DIM} vector로 날려줌
# 최종 tok_embedding(x)의 차원은 (bsz, seq_len, HID_DIM)
#
# pos도 x의 변환과 동일
# 어떠한 정보를 넣어줄 것이냐의 차이만 존재
# token information vs position information
#
# 왜 scale을 곱해주냐?
# 차차 생각해보자...

src
```

```python
src.shape # (batch, seq_len, hid_dim)
```

- MultiHeadAttention(self-attention)

```python
ENC_HEAD_DIM = HID_DIM // ENC_HEADS
ENC_HEAD_DIM # 256 // 8
```

```python
fc_q = nn.Linear(HID_DIM, HID_DIM)
fc_k = nn.Linear(HID_DIM, HID_DIM)
fc_v = nn.Linear(HID_DIM, HID_DIM)
fc_o = nn.Linear(HID_DIM, HID_DIM)
```

```python
query = key = value = src
```

```python
Q = fc_q(query)
K = fc_k(key)
V = fc_v(value)
```

```python
Q.shape, K.shape, V.shape # (batch_size, seg_len, hid_dim)
```

```python
print(Q.view(BATCH_SIZE, -1, ENC_HEADS, ENC_HEAD_DIM).shape)
print(Q.view(BATCH_SIZE, -1, ENC_HEADS, ENC_HEAD_DIM).permute(0, 2, 1, 3).shape)

# (batch_size, seq_len, n_heads, head_dim)
# (batch_size, n_heads, seq_len, head_dim)
```

```python
Q = Q.view(BATCH_SIZE, -1, ENC_HEADS, ENC_HEAD_DIM).permute(0, 2, 1, 3)
K = K.view(BATCH_SIZE, -1, ENC_HEADS, ENC_HEAD_DIM).permute(0, 2, 1, 3)
V = V.view(BATCH_SIZE, -1, ENC_HEADS, ENC_HEAD_DIM).permute(0, 2, 1, 3)
```

```python
Q.shape, K.shape
```

```python
energy = torch.matmul(Q, K.permute(0, 1, 3, 2)) / scale
```

```python
res = torch.matmul(dropout(attention), V)
```

```python
res = res.permute(0, 2, 1, 3).contiguous()
res.shape # (batch_size, seq_len, n_heads, hid_dim / n_heads)
```

```python
res = res.view(BATCH_SIZE, -1, HID_DIM)
res.shape # (batch_size, seq_len, hid_dim)
```

```python
res = fc_o(res)
res.shape # (batch_size, seq_len, hid_dim)
```

```python
_src = res
```
